```python
# coding: utf-8

"""
Crea una tabla con todos los csvs que se encuentran en ./accidents, selecciona 
solo las del estado y las caracteristicas indicadas en las variables 'estado' y 'dtypes'. 
Elimina aquellas que no tienen la posicion en la que ocurre la tormenta y pone las fechas en un formato adecuado. 
Finalmente vuelva el resultado en un archivo .csv
"""

import pandas as pd
import numpy as np
from functools import partial
from os import listdir
```

```python
''' seleccionamos los accidentes del estado de indiana'''
accidents_csv = []
for csv in listdir('../Data/Iniciales/accidents/'):
    df = pd.read_csv('../Data/Iniciales/accidents/' + csv)
    df.set_index = "INDEX"
    accidents_csv.append(df)

accidents = pd.concat(accidents_csv)
                          
accidents = accidents[accidents['STATE'] == 18]

''' Eliminamos las columnas que no sirven'''

useless_data = ['ALIGNMNT', 'ARR_HOUR', 'ARR_MIN', 'CF1', 'CF2', 'CF3',
                'CITY', 'COUNTY', 'C_M_ZONE', 'DAY_WEEK', 'FUNC_SYS',
                'HARM_EV', 'HIT_RUN', 'HOSP_HR', 'HOSP_MN', 'LGT_COND', 'MAN_COLL',
                'MILEPT', 'NHS', 'NOT_HOUR', 'NOT_MIN', 'NO_LANES', 'PAVE_TYP', 'PEDS',
                'PERMVIT', 'PERNOTMVIT', 'PERSONS', 'PROFILE', 'PVH_INVL', 'RAIL', 'RD_OWNER',
                'RELJCT1', 'RELJCT2', 'REL_JUNC', 'REL_ROAD', 'ROAD_FNC', 'ROUTE', 'RUR_URB',
                'SCH_BUS', 'SP_JUR', 'SP_LIMIT', 'ST_CASE', 'SUR_COND', 'TRAF_FLO', 'TRA_CONT',
                'TWAY_ID', 'TWAY_ID2', 'TYP_INT', 'T_CONT_F', 'VE_FORMS', 'VE_TOTAL', 'WEATHER',
                'WEATHER1', 'WEATHER2', 'WRK_ZONE']

accidents = accidents.drop(useless_data, axis=1)
accidents
```

```python
''' Guardamos el csv '''

accidents.to_csv("../Data/Seleccion/accidents.csv")
```
