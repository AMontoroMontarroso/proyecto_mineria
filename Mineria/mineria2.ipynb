```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.neighbors
from scipy import cluster
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../Data/mineria/zonas.csv')
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
df['CLASE'] = 0
```

```python
df_zona1 = df[df['ZONA'] == 1]
df_tormentas1 = df_zona1[['GRANIZO', 'LLUVIA', 'NIEVE', 'T. ELECTRICA', 'TORNADO', 'VIENTO', 'DURACION']]

min_max_scaler = MinMaxScaler()
norm_tormentas1 = min_max_scaler.fit_transform(df_tormentas1)

pca_estimator = PCA(n_components=2)
X_tormentas1 = pca_estimator.fit_transform(norm_tormentas1)
print pca_estimator.explained_variance_ratio_

plt.scatter(X_tormentas1[:,0], X_tormentas1[:,1])
plt.show()
```

```python
dist = sklearn.neighbors.DistanceMetric.get_metric('manhattan')
matsim = dist.pairwise(norm_tormentas1)
clusters = cluster.hierarchy.linkage(matsim, method='average')
cluster.hierarchy.dendrogram(clusters, color_threshold=13)
plt.show()
```

```python
# Probamos a hacer un scatter plot con los 
# datos del pca y los labels sacados del jerarquico

cut = 10
labels = cluster.hierarchy.fcluster(clusters, cut, criterion='distance')
plt.scatter(X_tormentas1[:,0], X_tormentas1[:,1], c=labels)
plt.show()
```
