```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/Transformacion/transformacion2.csv')
```

```python
plt.scatter(df['LON'], df['LAT'])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
init = 'k-means++' 
iterations = 10 
max_iter = 300 
tol = 1e-04


ks = []
for k in range(1, 51):
    km = KMeans(k, n_init = iterations, max_iter= max_iter, tol=tol, init=init)
    labels = km.fit_predict(df[['LON', 'LAT']])
    ks.append(km.inertia_)

plt.plot(ks)
plt.show()
```

```python
k = 15

km = KMeans(k, n_init = iterations, max_iter= max_iter, tol=tol, init=init)
labels = km.fit_predict(df[['LON', 'LAT']])

plt.scatter(df['LON'], df['LAT'], c = labels)
plt.show()
```

```python
df['ZONA'] = labels

df_zona1 = df[df['ZONA'] == 1]
df_tormentas1 = df_zona1[['GRANIZO', 'LLUVIA', 'NIEVE', 'T. ELECTRICA', 'TORNADO', 'VIENTO', 'DURACION']]
df_tormentas1
```

```python
import sklearn.neighbors
from scipy import cluster
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
norm_tormentas1 = min_max_scaler.fit_transform(df_tormentas1)

pca_estimator = PCA(n_components=2)
X_tormentas1 = pca_estimator.fit_transform(norm_tormentas1)
print pca_estimator.explained_variance_ratio_

plt.scatter(X_tormentas1[:,0], X_tormentas1[:,1])
plt.show()
```

```python
dist = sklearn.neighbors.DistanceMetric.get_metric('manhattan')
matsim = dist.pairwise(df_tormentas1)
clusters = cluster.hierarchy.linkage(matsim)
```
