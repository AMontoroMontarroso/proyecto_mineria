```python
import pandas as pd
from pandas import DataFrame
''' En esta parte preprocesamos los datos de los accidentes. En primer lugar
    Vamos a convertir las tablas de la fecha en fechas reales. Despues se van
    a detectar posibles valores nulos que se hayan podido causar
'''

accidents = pd.read_csv('../Data/Seleccion/accidents.csv')
```

```python
# transformamos los datos del año, mes y día a entero (antes estaba en float)
accidents.YEAR = accidents.YEAR.map(lambda x: int(x))
accidents.DAY = accidents.DAY.map(lambda x: int(x))
accidents.MONTH = accidents.MONTH.map(lambda x: int(x))
```

```python
# definimos la funcion que nos va a transformar a fecha

def to_date(accidents):
    result = []
    for i in range(len(accidents['YEAR'])):
        res = "{}/{}/{}".format(accidents['YEAR'][i], accidents['MONTH'][i], accidents['DAY'][i],)
        result.append(pd.to_datetime(res, format="%Y/%m/%d"))
    return result
```

```python
'''
Pasamos las fechas a datetime de pandas y luego lo insertamos en una nueva columna
Una vez hecho eso, se eliminan las columnas que no hagan falta
'''

res = to_date(accidents)

accidents['DATE'] = DataFrame(res)

accidents = accidents.drop(['HOUR', 'MINUTE', 'DAY', 'HOUR', 'MONTH', 'YEAR'], axis=1)
accidents
```

```python
# Vemos si algun valor es nulo. Como ninguno es nulo, en principio es correcto

accidents.isnull().any()
```

```python
# Dibujamos los resultados, no vaya a ser que esto esté mal
import matplotlib.pyplot as plt

plt.scatter(accidents['LATITUDE'], accidents['LONGITUD'])
plt.show()
```

```python
# Hay unas fechas mal metidas, así que vamos a eliminarlas

accidents = accidents[accidents["LATITUDE"] > 20]
accidents = accidents[accidents["LATITUDE"] < 40]

plt.scatter(accidents['LATITUDE'], accidents['LONGITUD'])
plt.show()

# Ahora funciona
```
